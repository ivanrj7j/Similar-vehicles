# Evaluating models

## Loading models

```python
import tensorflow as tf
from tensorflow import keras
from keras.models import load_model
```

```python
v1 = load_model('modelV1.h5')
v2 = load_model('modelV2.h5')
```

```python
from keras.utils import image_dataset_from_directory
```

```python
testData = image_dataset_from_directory('augmented data/test', label_mode='categorical')
```

## Evaluating Model

```python
from keras.metrics import Accuracy
```

```python
v1_accuracy = Accuracy()

for i,batch in enumerate(testData):
    x, y = batch
    yhat = v1.predict(x)
    v1_accuracy.update_state(y, yhat)

```

```python
v1_accuracy.result()
```

```python
v2_accuracy = Accuracy()

for i,batch in enumerate(testData):
    x, y = batch
    yhat = v2.predict(x)
    v2_accuracy.update_state(y, yhat)

v2_accuracy.result()
```

```python
v2_accuracy.result().numpy(), v1_accuracy.result()
```

```python
from PIL import Image
```

```python
testData.class_names
```

```python
def predict(image:str, model:keras.models.Model):
    img = tf.convert_to_tensor(Image.open(image).resize((256, 256)))
    prediction = model.predict(tf.expand_dims(img, 0), verbose=False)
    return testData.class_names[int(tf.argmax(prediction, 1).numpy())]
```

```python
target = 'partitioned data/test/bikes/image.jpeg'
predV1 = predict(target, v1)
predV2 =  predict(target, v2)

predV1, predV2
```

For some reason v1 accuracy is 56% while v2 accuracy is 0%, but the v2 model is performing really well compared to v1
